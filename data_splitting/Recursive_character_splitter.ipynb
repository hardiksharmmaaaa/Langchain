```python
## Loading the pdf 

from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader("Data_science_research.pdf")
```

```python
Text_document=loader.load()
Text_document
```

## How to Recursivelt split text by characters

```python
# We require the library langchain-text-splitters

from langchain_text_splitters import RecursiveCharacterTextSplitter  # responsible for recursively split the text by characters

text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_doc=text_splitter.split_documents(Text_document)
final_doc
```

```python
print(final_doc[0])
print("-----------------------------------------------------Page break-------------------------------------------------------------------------------")
print(final_doc[1])
```

```python
## speech loading 

# Loadking the text file 

from langchain_community.document_loaders import TextLoader
loader=TextLoader("speech.txt")
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter  # responsible for recursively split the text by characters
speech=""
with open("speech.txt") as f:
    speech=f.read()
speech
```

```python

text_splitter=RecursiveCharacterTextSplitter(chunk_size=50,chunk_overlap=10)
final_doc=text_splitter.create_documents([speech])
print(final_doc)
```

```python
print(final_doc[0])
print(final_doc[1])
print(final_doc[2])
```
