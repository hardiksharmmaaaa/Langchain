### How to Split JSON Data 

```python
import json
import requests

json_data=requests.get("https://api.smith.langchain.com/openapi.json").json()   # My json Data 
```

```python
json_data
```

```python
from langchain_text_splitters import RecursiveJsonSplitter
json_splitter=RecursiveJsonSplitter(max_chunk_size=300)
json_chunks=json_splitter.split_json(json_data)
json_chunks
```

```python
for i in json_chunks[:3]:
    print(i,end="\n")
```

```python
## The Splitter can also output in the form of document

docs=json_splitter.create_documents(texts=[json_data])
for doc in docs[:3]:
    print(docs,end="\n")
```
